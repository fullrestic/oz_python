#### 프로젝트: 붕어빵 자동 판매 시스템


당신은 붕어빵 가게의 주인입니다.

손님들이 붕어빵을 주문하면 이를 자동으로 처리하고, 재고를 관리하는 시스템을 만들고자 합니다.

또한, 관리자 모드를 통해 재고를 추가할 수 있는 기능을 구현해야 합니다.


__[문제 1: 붕어빵 재고 확인]__

붕어빵 가게의 현재 재고를 확인하는 기능을 구현하세요.

요구사항:
- 현재 붕어빵의 재고를 딕셔너리로 관리합니다.
- 딕셔너리에는 붕어빵의 종류와 각각의 재고 개수가 저장됩니다.
- 현재 재고를 화면에 출력하세요.

붕어빵 재고
- 팥붕어빵 : 10개
- 슈크림붕어빵 : 8개
- 초코붕어빵 : 5개

```python
stock_list = {'팥붕어빵' : 10, '슈크림붕어빵' : 8, '초코붕어빵' : 5}

print('현재 붕어빵 재고')
print('------------')

for i in stock_list :
    print(f'{i} {stock_list[i]}개')
```

__[문제 2: 붕어빵 주문 받기]__

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
order_list = {}

fish_type = input('주문할 붕어빵의 종류(팥붕어빵, 슈크림붕어빵, 초코붕어빵)를 입력해주세요 : ')
fish_order = int(input('주문할 붕어빵 개수를 입력해주세요 : '))
order_list.update({fish_type : fish_order})

print('주문 내역')
print('------------')
print(f'{fish_type} {order_list[fish_type]}개')
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
fish_sell = 0

if order_list[fish_type] > stock_list[fish_type] :
    print('재고가 부족합니다.\n')
else :
    stock_list[fish_type] -= order_list[fish_type]
    fish_sell = order_list[fish_type]

print('판매 내역')
print('------------')
print(f'{fish_type} {fish_sell}개')
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
while True :
    fish_type = input('주문할 붕어빵의 종류(팥붕어빵, 슈크림붕어빵, 초코붕어빵)를 입력해주세요 : ')
    if fish_type == '종료' : break

    fish_order = int(input('주문할 붕어빵 개수를 입력해주세요 : '))

    if fish_order > stock_list[fish_type] :
        print('재고가 부족합니다.\n')
    else :
        stock_list[fish_type] -= fish_order

print('\n현재 붕어빵 재고')
print('------------')

for i in stock_list :
    print(f'{i} {stock_list[i]}개')
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
while True :
    update_type = input('추가 붕어빵 종류 : ')
    if update_type == '종료' : break

    update_number = int(input('추가 수량 : '))

    stock_list[update_type] += update_number

print('\n현재 붕어빵 재고')
print('------------')

for i in stock_list :
    print(f'{i} {stock_list[i]}개')
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
fish_price = {'팥붕어빵' : 1000, '슈크림붕어빵' : 1200, '초코붕어빵' : 1500}
fish_sold = {'팥붕어빵' : 0, '슈크림붕어빵' : 0, '초코붕어빵' : 0}
total_sales = 0

while True :
    fish_type = input('주문할 붕어빵의 종류(팥붕어빵, 슈크림붕어빵, 초코붕어빵)를 입력해주세요 : ')
    if fish_type == '종료' : break

    fish_order = int(input('주문할 붕어빵 개수를 입력해주세요 : '))

    if fish_order > stock_list[fish_type] :
        print('재고가 부족합니다.\n')
    else :
        stock_list[fish_type] -= fish_order
        fish_sold[fish_type] += fish_order

print('\n총 매출')
print('------------')

for i in fish_price :
    print(f'{i} {fish_sold[i]}개')
    total_sales += fish_price[i] * fish_sold[i]

print('------------')
print(total_sales, '원')
```
