- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
try : 
    num_list = []
    while True : 
        user_input = input('숫자를 입력하세요 : ')
        if user_input == 'done' : break
        num_list.append(int(user_input))
except ValueError : 
    print(f'잘못된 값 : {user_input}')
    print('숫자를 입력해주세요.')
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time as t

def timer_decorator(function) : 
    def wrapper() : 
        start = t.time()
        function()
        end = t.time()

        print(f'함수 실행 시간 : {end - start :.2f}')
    return wrapper

@timer_decorator
def funct() :
    cnt = 3000
    while cnt : 
        print('함수 실행', end = ' ')
        cnt -= 1
    print()
funct()
````

````python
# mymath.py

def tri_area(b, h) :
    return b * h * 0.5

def circ_area(r) :
    return r * r * 3.14

def rect_area(b, h) :
    return b * h
````

````python
# mymath_main.py

import mymath as mm

b, h = map(int, input('삼각형의 밑변과 높이를 입력하세요 : ').split())
print('삼각형의 넓이 : ', mm.tri_area(b, h))

r = int(input('원의 반지름을 입력하세요 : '))
print('원의 넓이 : ', mm.circ_area(r))

rb, rh = map(int, input('사각형의 밑변과 높이를 입력하세요 : ').split())
print('사각형의 넓이 : ', mm.rect_area(rb, rh))
````
